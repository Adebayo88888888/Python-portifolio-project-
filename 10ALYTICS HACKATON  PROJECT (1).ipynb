# CORRELATION TESTING

```python
# scatter plot with year and disease
plt.scatter(x=df['Poisonings'], y=df["Nutritional deficiencies"])
plt.title("YEARLY TRENDS FOR DISEASES")
plt.xlabel('Poisonings')
plt.ylabel("Nutritional deficiencies")
plt.show()
```

```python
# USING SEABORN FOR ANALYSIS
sns.regplot(x='Poisonings', y="Nutritional deficiencies" ,data = df,  scatter_kws={'color':'blue'}, line_kws={'color':'black'})
plt.savefig('africa_data_2.png')
plt.show()
```

```python
correlation_matrix= df.corr(method= 'pearson')
```

```python
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot= True )
plt.title("CORRELATION MATRIC FOR NUMERIC FEATURE")
plt.xlabel('HEALTH FEATURE')
plt.ylabel('HEALTH FEATURE')
# save plot
plt.savefig.('africa_data.png')
plt.show()
```
